```python
import pandas as pd  
import numpy as np 
import jieba
from snownlp import SnowNLP
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_excel('./结果/情感分析结果.xlsx')
df.head(1)
```

```python
sentiments=df['情感值']

#各情感分数段出现频率柱状图
plt.hist(sentiments, bins=np.arange(0, 1, 0.01), facecolor='g')
plt.xlabel('Sentiments Probability')
plt.ylabel('Quantity')
plt.title('Analysis of Sentiments')
plt.savefig('./结果/各情感分数段出现频率柱状图.png')
plt.show()
```

```python
#情感波动分析
sentimentslist=df['情感值']
plt.plot(np.arange(0, 9646, 1), sentimentslist, 'b-')
plt.xlabel('Number')
plt.ylabel('sentimentslist')
plt.title('Analysis of Sentiments')
plt.savefig('./结果/情感波动分析图.png')
plt.show()
```

```python
#极性占比图
grp=df['情感极性'].value_counts()
grp.plot.pie(y='情感极性',autopct='%.2f%%') #画饼图
plt.title('情感极性占比图')
plt.savefig('./结果/情感极性占比图.png')
plt.rcParams['font.family'] = ['Heiti TC']
plt.show()
```
