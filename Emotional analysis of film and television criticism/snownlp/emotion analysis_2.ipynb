```python
from snownlp import SnowNLP
import codecs
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv  
```

```python
sentimentslist = []   # 创建一个空列表，用于存储情感分值  
count = 0  # 初始化一个计数器，用于记录已经处理的行数

#从一个CSV文件中读取前2000行的文本，并使用SnowNLP库对每一行进行情感分析，然后将结果存储在一个列表中。同时，它还提供了错误处理机制，以确保在发生异常时能够优雅地处理。
# 用了with语句来打开文件。使用with语句可以确保文件在操作完成后自动关闭，即使发生异常也是如此。
with open("./data/豆瓣肖申克的救赎.csv", "r", encoding="utf-8") as source:
    for line in source:   #逐行读取文件的内容，并将每一行的内容存储在变量line中。
        if count < 2000:  # 仅处理前2000行
            try:   #尝试对line进行情感分析
                s = SnowNLP(line.strip()) #使用SnowNLP库创建一个对象s，并传入处理后的line（使用strip()方法去除可能的换行符和空白字符）。
                sentimentslist.append(s.sentiments)  #将s对象的sentiments属性（情感分值）添加到sentimentslist列表中#
                print(s.sentiments)  # 打印情感分值，可选
            except Exception as e:  #如果在尝试过程中发生异常（比如SnowNLP无法处理某些特殊字符或格式），则捕获该异常
                print(f"Error processing line: {line} with error: {e}")  #打印出出错的行和错误信息。
            count += 1  # 无论是否发生异常，都会更新计数器。
        else:
            break  # 如果达到2000行，退出循环
```

```python
# 假设情感分值阈值  
POSITIVE_THRESHOLD = 0.6  
NEGATIVE_THRESHOLD = 0.4  
  
# 初始化情感分值和情感极性列表  
sentiments_scores = []  
sentiments_polarities = []  
  
# 计数器  
count = 0  
  
# 使用with语句来打开文件  
with open("./data/豆瓣肖申克的救赎.csv", "r", encoding="utf-8") as source:  
    reader = csv.reader(source)  
    for line in reader:  
        if count < 2000:  
            try:  
                # 假设第一列是包含评论的列  
                comment = line[0].strip()  
                s = SnowNLP(comment)  
                score = s.sentiments  
                  
                # 根据情感分值设置情感极性  
                if score >= POSITIVE_THRESHOLD:  
                    polarity = '积极'  
                elif score <= NEGATIVE_THRESHOLD:  
                    polarity = '消极'  
                else:  
                    polarity = '中级'  
                  
                sentiments_scores.append(score)  
                sentiments_polarities.append(polarity)  
                  
                # 可选：打印情感分值和情感极性  
                # print(f"情感分值: {score}, 情感极性: {polarity}")  
                  
            except Exception as e:  
                print(f"Error processing line: {line} with error: {e}")  
              
            count += 1  
        else:  
            break  
  
# 创建DataFrame  
df = pd.DataFrame({  
    '情感分值': sentiments_scores,  
    '情感极性': sentiments_polarities  
})  
  
# 将DataFrame保存到Excel文件  
df.to_excel('sentiments_output.xlsx', index=False)  
  
print('情感分析结果已保存到sentiments_output.xlsx')
```

```python
import mysql.connector  
from snownlp import SnowNLP  
import csv  
  
# MySQL数据库连接信息  
config = {  
    'host': 'localhost',  
    'user': 'root',  
    'password': '123456',  
    'database': 'doubanmovie',  
    'port': 3306,  
    'charset': 'utf8mb4'  
}  
  
# 创建数据库连接  
cnx = mysql.connector.connect(**config)  
cursor = cnx.cursor()  
  
# 假设我们有一个表sentiment_analysis，有两个字段：sentiment_score和sentiment_polarity  
# 如果表不存在，则创建它  
create_table_query = """  
CREATE TABLE IF NOT EXISTS sentiment_analysis (  
    id INT AUTO_INCREMENT PRIMARY KEY,  
    sentiment_score FLOAT NOT NULL,  
    sentiment_polarity VARCHAR(255) NOT NULL  
)  
"""  
cursor.execute(create_table_query)  
cnx.commit()  
  
# 初始化情感分值和情感极性列表  
sentiments_scores = []  
sentiments_polarities = []  
  
# 计数器  
count = 0  
  
# 使用with语句来打开文件  
with open("./data/豆瓣肖申克的救赎.csv", "r", encoding="utf-8") as source:  
    reader = csv.reader(source)  
    for line in reader:  
        if count < 2000:  
            try:  
                # 假设第一列是包含评论的列  
                comment = line[0].strip()  
                s = SnowNLP(comment)  
                score = s.sentiments  
                  
                # 根据情感分值设置情感极性  
                if score >= 0.6:  
                    polarity = '积极'  
                elif score <= 0.4:  
                    polarity = '消极'  
                else:  
                    polarity = '中级'  
                  
                sentiments_scores.append(score)  
                sentiments_polarities.append(polarity)  
                  
            except Exception as e:  
                print(f"Error processing line: {line} with error: {e}")  
              
            count += 1  
        else:  
            break  
  
# 插入数据到MySQL表  
insert_query = """  
INSERT INTO sentiment_analysis (sentiment_score, sentiment_polarity)  
VALUES (%s, %s)  
"""  
  
# 使用executemany一次性插入多行数据  
data_to_insert = list(zip(sentiments_scores, sentiments_polarities))  
cursor.executemany(insert_query, data_to_insert)  
cnx.commit()  
  
# 关闭数据库连接  
cursor.close()  
cnx.close()  
  
print('情感分析结果已保存到MySQL数据库。')
```

```python
# 打开文件并一次性读取所有行
with open("./data/豆瓣肖申克的救赎.csv", "r", encoding="utf-8") as source:
    lines = source.readlines()

# 限制仅处理前2000行
lines = lines[:2000]

#对每行文本进行情感分析
sentimentslist = []  #初始化一个空列表sentimentslist，用于存储每行文本的情感分值
for i in lines:  #使用for循环遍历lines列表中的每一行文本
    s = SnowNLP(i)  #对于每一行文本i，使用SnowNLP库创建一个对象s，并传入该文本
    sentimentslist.append(s.sentiments)  #调用s.sentiments获取该文本的情感分值，并将其添加到sentimentslist列表中

# 情感波动分析
result = [x - 0.5 for x in sentimentslist]  # 使用列表推导式创建一个新列表result,对于sentimentslist列表中的每个情感分值x，计算x - 0.5，并将结果添加到result列表中

# 计算情感分类
categories = {'消极': 0, '一般': 0, '积极': 0}  #初始化一个字典categories，用于存储三种情感分类（消极、一般、积极）的计数
for sentiment in sentimentslist:  #使用for循环遍历sentimentslist列表中的每个情感分值
    if sentiment < 0.4:
        categories['消极'] += 1  #如果情感分值小于0.4，则认为是消极情感
    elif sentiment < 0.6:
        categories['一般'] += 1  #如果情感分值在[0.4, 0.6)之间，则认为是一般情感
    else:
        categories['积极'] += 1  #如果情感分值大于或等于0.6，则认为是积极情感
```

```python
# 数据准备
sizes = [categories['消极'], categories['一般'], categories['积极']]  #一个列表，包含每种情感分类的计数
labels = ['Negative', 'Neutral', 'Positive']  #一个列表，包含每种情感分类的标签
colors = ['red', 'yellow', 'green']  #一个列表，包含每种情感分类在饼图中对应的颜色

# 绘制饼图
plt.figure(figsize=(8, 6))  # 创建一个新的图形，设置其大小为8x6英寸
# sizes: 数据大小  
# labels: 数据标签  
# colors: 每种数据对应的颜色  
# autopct: 自动显示百分比，格式化为带有一位小数的百分比  
# startangle: 饼图的起始角度  
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)  
plt.title('Sentiment Analysis Results')  # 设置饼图的标题 
plt.axis('equal')   #确保饼图是圆形的，而不是椭圆形的

# 可视化画图，绘制情感波动线图
plt.figure(figsize=(10, 6))  #创建一个新的图形，设置其大小为10x6英寸
plt.plot(np.arange(len(result)), result, 'k-')  # 绘制线图  
plt.xlabel('Number')  # 设置x轴的标签 
plt.ylabel('Sentiment')  # 设置y轴的标签 
plt.title('Analysis of Sentiments')  # 设置图表的标题 

plt.show()  # 显示图表
```
